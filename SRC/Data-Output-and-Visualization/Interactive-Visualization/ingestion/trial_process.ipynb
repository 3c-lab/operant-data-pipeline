```python
def serialize_timestamps(string_of_timestamps):
    timestamp_list = literal_eval(string_of_timestamps)

    while type(timestamp_list[-1]) == list:
        timestamp_list.pop()
    
    if sum(timestamp_list) == 0:
        return "no timestamps"
    
    while timestamp_list[-1] == 0:
        timestamp_list.pop()
    
    lst_strings = [str(i) for i in timestamp_list]
    return " ".join(lst_strings)
```

```python
lst2 = "[0,1,1,0, [1,2,3]]"
serialize_timestamps(lst2)
```

```python
all_lengths = []

for t in ["active_timestamps", "inactive_timestamps", "reward_timestamps", "timeout_timestamps"]:
    for i in range(len(test_df)):
        test_row = test_df.iloc[i]
        vals = test_row.get(t)
        output_string = serialize_timestamps(vals)
        all_lengths.append(len(output_string))
```
