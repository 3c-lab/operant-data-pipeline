## SHA

```python
filepath = '/Users/yunyihuang/Desktop/COCAINE/SHA'
files = [i for i in sorted(os.listdir(filepath)) if i != '.DS_Store']
files
```

```python
len(files)
```

```python
# calculation for 1 file
test = files[0]
df = cleanup(os.path.join(filepath, test))
dff = calculations_single(df)
trial_id = re.findall(r'((?:SHA|LGA)[0-9]+)',test.upper())[0]
dff['trial_id'] = [trial_id] * len(dff)
dff
```

```python
calcs = []
for i in tqdm(range(len(files))):
    f = files[i]
    df = cleanup(os.path.join(filepath, f))
    calc = calculations_single(df)
    trial_id = re.findall(r'((?:SHA|LGA)[0-9]+)',f.upper())[0]
    calc['trial_id'] = [trial_id] * len(calc)
    calcs.append(calc)
```

```python
# calcs 0-311
# calcs 313-466
# calcs 468-495
# calcs 497-534
all_calcs = calcs + calcs2 + calcs3 + calcs4
all_calcs
```

```python
df_rawcalcs = pd.concat(all_calcs, ignore_index=True)
df_rawcalcs.head()
```

```python
df_rawcalcs.shape
```

```python
df_rawcalcs.Subject.value_counts()
```

```python
rows = []
subs = [i for i in df_rawcalcs.Subject.unique() if i != '0']
for s in subs:
    temp = df_rawcalcs[df_rawcalcs.Subject == s]
    temp = temp.reset_index(drop=True)
    temp.sort_values(['Subject','trial_id'],inplace=True)
    
    temp = temp.set_index('trial_id')
    temp_row = temp.iloc[:,1:].unstack().to_frame().sort_index(level=1).T
    temp_row.insert(loc=0, column='Subject', value=temp.Subject.unique())
    temp_row = temp_row.reset_index(drop=True)
    temp_row.columns = [col[0] + '_' + col[1] if len(col[1]) > 0 else col[0] for col in temp_row.columns]
    temp_row = temp_row.loc[:,~temp_row.columns.duplicated()].copy()
    rows.append(temp_row)
```

```python
len(rows)
```

```python
out = pd.concat(rows,ignore_index=True)
out
```

```python
out.to_csv('burst_data_COCAINE_SHA.csv')
```
